This Python code demonstrates a simple linear regression analysis.
We use a dataset representing house prices and their corresponding sizes.
The aim is to plot the data along with the regression line represented by the equation Y = 30 + 0.18X.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
X = np.array([1000, 1500, 1200, 1800, 1300, 1600, 1100, 1550, 1700, 1400])  # House Size
Y = np.array([200, 300, 250, 350, 280, 320, 230, 310, 340, 290])  # House Price

# Regression equation
def regression_model(X):
    return 30 + 0.18 * X

# Plotting the data
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, regression_model(X), color='red', label='Regression Line: Y = 30 + 0.18X')

plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($K)')
plt.title('House Price Prediction Based on Size')
plt.legend()
plt.show()
```
