# Step 1: Setup and Data Preparation

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset loading and preparation steps
# Assume X and y are loaded, where X is the feature matrix and y is the label vector
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

# Step 2: Implementing Logistic Regression

```python
# Sigmoid function to map predictions to probabilities.
def sigmoid(z):
    # Apply the sigmoid function to any value 'z' to get a value between 0 and 1.
    return 1 / (1 + np.exp(-z))

# Function to compute the cost of a given solution (characterized by 'theta').
def compute_cost(X, y, theta):
    # Number of training examples.
    m = len(y)
    # Compute the predicted probabilities for all training examples.
    predictions = sigmoid(np.dot(X, theta))
    # Calculate the cost using the logistic regression cost formula.
    cost = -(1/m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    # Return the cost.
    return cost

# Function to perform gradient descent to learn the optimal 'theta' parameters.
def gradient_descent(X, y, theta, alpha, iterations):
    # Number of training examples.
    m = len(y)
    # Array to store the cost at each iteration.
    cost_history = []

    # Perform the gradient descent algorithm for a specified number of iterations.
    for i in range(iterations):
        # Compute the gradient (partial derivatives) of the cost function.
        gradient = (1/m) * np.dot(X.T, (sigmoid(np.dot(X, theta)) - y))
        # Update 'theta' by taking a step against the gradient, scaled by the learning rate.
        theta -= alpha * gradient
        # Record the cost with the updated 'theta' for plotting.
        cost_history.append(compute_cost(X, y, theta))

    # Return the learned parameters and the history of cost values.
    return theta, cost_history
```

# Training the Model

```python
# Initializations for logistic regression training.
# 'theta_initial' is the starting point for the optimization and is initially set to zeros.
# This vector has the same number of elements as features in the dataset.
theta_initial = np.zeros(X.shape[1])

# 'alpha' is the learning rate, which controls the size of the update step on each iteration of gradient descent.
alpha = 0.01

# 'iterations' is the number of times the gradient descent algorithm will update the 'theta' parameters.
iterations = 1000

# Training the logistic regression model using the gradient descent optimization algorithm.
# 'theta_optimal' will hold the final optimized values of the parameters after gradient descent has run.
# 'cost_history' will contain the value of the cost function after each iteration, showing how it changes over time.
theta_optimal, cost_history = gradient_descent(X, y, theta_initial, alpha, iterations)

# Plotting the cost function over iterations.
# A plot of 'cost_history' helps us understand if the cost is decreasing over time, 
# which indicates that the gradient descent is working correctly.
plt.plot(cost_history)
plt.xlabel('Iteration')  # X-axis label representing the iteration number.
plt.ylabel('Cost')       # Y-axis label representing the cost value.
plt.title('Cost Function Over Iterations')  # Title of the plot.
plt.show()  # Display the plot.
```
